```python
import torch

from dawnet.inspector import LLMInspector
from dawnet.prompts import get_words
```

```python
word_lists = get_words(with_category=False, n_per_category=3)
print(word_lists)
```

```python
model_id = "google/gemma-7b"
# model_id = "google/gemma-2-2b-it"
# model_id = "google/gemma-3-27b-it"
# model_id = "openai/gpt-oss-120b"
# model_id = "Qwen/Qwen3-4B"
insp = LLMInspector.from_hf(model_id)
```

```python
for word in word_lists:
  toks = insp.encode(word)
  if "gemma" in insp.model_id:
    toks = toks[:,1:]
  output = insp.infer(toks)
  print(f"'{word}' - '{output.argmax()[1]}'")
```

```python
longer_sequence = "When John and Mary went to the store, John gave the bag to"
toks = insp.encode(longer_sequence)
if "gemma" in insp.model_id:
  toks = toks[:,1:]
output = insp.infer(toks)
print(f"'{output.argmax()[1]}'")
```
