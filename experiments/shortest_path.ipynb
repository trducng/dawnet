```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

from dawnet.inspector import LLMInspector

torch.set_grad_enabled(False)
```

```python
model_id = "Qwen/Qwen3-4B-Thinking-2507"
device = torch.device("mps")
insp = LLMInspector.from_hf(model_id).to(device=device)
```

```python
prompt = insp.encode("What is the largest city in Europe?", chat=True)
print(insp.tokenizer.decode(prompt[0]))
```

```python
tokens = insp.generate(prompt, max_new_tokens=4096)
print(tokens)
```

```python
tokens.view_nb()
```

```python
new_prompt_1 = tokens.drop_tokens_randomly(pdrop=0.5, indices=list(range(213)), truncate=True)
print(new_prompt_1)
```

```python
out1 = insp.generate(new_prompt_1.tensor, max_new_tokens=4096)
print(out1)
```
