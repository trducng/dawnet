# Draw history progress from checkpoint

```python
# path to checkpoint
CHECKPOINT_PATH = ''

# the value for $CUDA_VISIBLE_DEVICES, '' means using cpu
CUDA_DEVICE = ''
```

```python
%matplotlib inline
import os

import torch

from dawnet.diagnose.vis import draw_history

os.environ['CUDA_VISIBLE_DEVICES'] = CUDA_DEVICE
if CHECKPOINT_PATH.strip() == '':
    raise AttributeError('please supply `CHECKPOINT_PATH`')


print('Loading state...')
state = torch.load(CHECKPOINT_PATH, map_location=lambda storage, loc: storage)
history = state['history']
print('State loaded')
```

## General plotting

```python
%matplotlib notebook

draw_history(history)
```

## Specific plotting

```python
%matplotlib notebook

draw_history(history, 'loss')
```

```python
%matplotlib notebook

draw_history(history, 'lr')
```
